# In this project, I will Collect data about people in all countries of the world ,and i will put each table in its own file
## Data from this website "https://www.worldometers.info/"


```python
# import modules i need it 

from os import listdir
from pandas import  read_csv
import requests
from bs4 import BeautifulSoup
```

```python
request=requests.get("https://www.worldometers.info/population/") #requst the Page
soup=BeautifulSoup(request.text,"html.parser")



data=soup.find("div",attrs={"class":"content-inner"}).find_all("ul")[8].find_all("li") #requst the tapul that has the name and the linke for evry country 
bass_url="https://www.worldometers.info"



for i in range(len(data)):
    country_url=data[i].find("a").attrs['href'] # requst the country url
    country_name=data[i].text # requst the country name
    

    request_padg=requests.get(bass_url+country_url) # get into the link to requst the data about each country
    padg_data=BeautifulSoup(request_padg.text,"html.parser")

    
    
    
    with open (f"Folder files\\{country_name.lower()}.csv", mode='w', encoding="utf-8") as file :
        ## the titel at the csv 
        file.write("Year,Population,Yearly_%_Change,Yearly_Change,Migrants_net,Median_Age\n")
        
        # get the valy 
        try:
            tabul=padg_data.find("div",attrs={"class":"table-responsive"}).find("tbody").find_all("tr")
        except:
            tabul=None
        
        if tabul != None:
            for vv in range(len(tabul)):
                
                valu_rov=tabul[vv].find_all("td")
                ## ---------------------------------------
                try:
                    Year = valu_rov[0].text
                except:
                    Year=" "

                ## ---------------------------------
                try:
                    Population = valu_rov[1].text
                except:
                    Population=" "
                ## --------------------------------
                try:
                    Yearly_1_Change =valu_rov[2].text
                except:
                    Yearly_1_Change=" "
                ## ------------------------------------
                try:
                    Yearly_Change =valu_rov[3].text
                except:
                    Yearly_Change=" "
                ## -------------------------------
                try:
                    Migrants_net=valu_rov[4].text
                except:
                    Migrants_net=" "
                ## --------------------------------
                try:
                    Median_Age =valu_rov[5].text
                except:
                    Median_Age=" "
                ## -------------------------------
                

                file.write(f"\"{Year}\",\"{Population}\",\"{Yearly_1_Change}\",\"{Yearly_Change}\",\"{Migrants_net}\",\"{Median_Age}\"\n") # add those valuse in the file 
        



    

```
